#### 训练数据分析

```python
train_data = pd.read_csv('./data/train.csv')
train_data.head(10)
```

```python
train_data.shape
```

```python
train_data = train_data.dropna(thresh=train_data.shape[0] * 0.5, axis = 1)
train_data.head(10)
```

```python
print(train_data['Open'].value_counts())

# 营业中
open_store = train_data[train_data['Open'] == 1]

print(open_store.count())

saled_store = open_store[open_store['Sales'] > 0]

print(saled_store.count())

data = pd.merge(data_store, saled_store, on='Store')
data.head()
```
